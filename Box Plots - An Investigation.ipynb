# Box Plots

In this notebook we will examine Box Plots, also known as Box and Whisker Diagrams, and their uses in Data Analytics.  

The Box Plot was invented by John Tukey, an American mathematicion, in 1969.  Tukey was awarded the National Medal of Science and was also awarded the Institute of Electrical and Electronics Engineers (IEEE)Medal of Honor in 1982 "For his contributions to the spectral analysis of random processes and the fast Fourier transform (FFT) algorithm."  

A Box Plot is the visual representation of the statistical five number summary of a given data set.  A five number summary includes th Minimum, First Quartile (Q1), Median (or Second Quartile, Q2), Third Quartile (Q3) and the Maximum. It can also tell you if your data is symmetrical, how tightly your data is grouped, and if and how your data is skewed.

A **Percentile** is a measure of relative standing within a dataset.  If we were to say, for example, that a value was in the 86th Percentile, we would be saying that 86% of the other values in the dataset are below that value.

A **Quartile** is an extension of percentile in that they mark three points in the percentile range. ie. Q1 is the 25th Percentile, Q2 is the median or 50th percentile and Q3 is the 75th percentile.

### The Elements

The Median, Q2/50th percentile, is the middle value of the dataset.  The first Quartile, Q1/25th percentile, is the middle value between the smallest (different to the 'minimum')and the median.  The third Quartile, Q3/75th percentile, is the middle value between the median and the largest number (different to the 'maximum').  The Interquartile range(IQR) is the distance from Q1 to Q3.

The maximum is then calulated by adding 1.5 times IQR to Q3 and the minimum is calculated by subtracting 1.5 times IQR from Q1.  This means that there may be numbers in the data set that are smaller than the 'minimum' and there may be number larger than the 'maximum'.  These values are known as 'Outliers'.

As a simple example to illustrate the five number summary we can use the following numbers:

1, 6, 10, 11, 14, 25, 32, 48, 50, 71,100

Q2 or Median = 25

Q1 = 10.5

Q3 = 49

IQR = 38.5

'Minimum' = Q1 - 1.5*IQR is -47.25 so that defaults to 1 as the 'minimum' as 1 is the smallest number in the data set and we don't go beyond the smallest.

'Maximum' = Q3 + 1.5*IQR = 82.75.  This means that the largest value in the data set, 100, is an Outlier

As the IQR is calculated by taking the middle number from the smallest to the median and the middle number from the median to the largest, it therefore follows that the IQR will always contain 50% of the values in a data set.

### Similarities with the Bell Curve for Normal Distribution
It is similar to a bell curve in that it provides the user with information on how the values in the data set are dispersed and it allows for outliers to be indentified without skewing the data results.
As normal distributions follow the 68-95-99.7 rule of 3 standard distributions meaning that 68% of values in a dataset fall within one standard distribution of the mean, 95% of values will fall within two stanard distributions of the mean and 99.7% of values will fall within 3 standard distributions of the mean.  This is best illustrated by using a bell curve displaying the density with which the values cluster around the mean.

A box plot provides similar information though calculated a little differently.  In a box plot, 50% of the values will fall in the IQR or the Box, with 24.65% on either side of the box by means of 'whiskers' to the 'minimum' and 'maximum' respectively, and 0.7% then accounting for outliers outliers.

## Graphing Box Plots using Python

In order to illustrate the graphing of a box plot and and how it can be used for comparative analysis we will use selected historical data from the **Tour de France**, a professional bicycle race in France.  There are 104 rows of values in the data set representing each of 104 races that have taken place to date.  The pace of each race is the variable to be graphed in the and for comparative purposes we will show the variance of distributions depending on the nationality of the winner.

```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
df = pd.read_csv("2018_historical_guide (5).csv") # import Tour de France historical dataset
```

```python
df
```

```python
# Create a Box Plot of the pace of every Tour de France to date
%matplotlib inline
sns.boxplot(x="Pace", data=df)
```

```python
df.median()
```

We can see from the Box Plot that the median or Q2 falls just short of 35.5 kph.  This means that 50% of all races were at a pace lower than 35.419kph and 50% were at a pace faster than 35.419kph.  The right section of the box is smaller than the left side and the whisker extending to the maximum is shorter than the whisker extending to the minimum and what this tells us is that there is less variance in speeds greater than the median than variance in speeds less than the median.  This would be in keeping with backround knowledge of the data, in that, as with all speed based sports, variances in pace become less and less as the sport continues due to a "plateauing" as the brink of human ability is reached.


```python
# Create a normal distribution graph to complare with box plot
%matplotlib inline
sns.distplot(df['Pace'])
```

```python
# Create a Standard Distribution graph of the pace of every Tourde France to date
df.loc[:,"Pace"].std()
```

```python
# Create a Box Plot of the pace of every Tour de France to date seperated by nationality of winner
sns.boxplot(x="Pace", y = "Country", data=df)
```
